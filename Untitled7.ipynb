# Código

- Criando uma constante PI = 3.141592...
- Criando um vetor para representar os valores do eixo do tempo
- Criando os valores de tensão do eixo a, b e c. Esses valores recebem o alias de za, zb e zc.

Obs: a velocidade angular "w" para esse exemplo recebe o valor de 1, mas em condições normais o valor dela é w = 2 * PI * frequencia_rede

```python
I = - 2 * PI / w
F = 2 * PI / w
t = np.arange(I , F + 0.000001, 0.000001)
za = np.cos(w * t)
zb = np.cos(w * t + (2 * PI / 3))
zc = np.cos(w * t - (2 * PI / 3))
```

```python
fig = plt.figure(figsize=(20, 6))

plt.plot(t, za, '-r')
plt.plot(t, zb, '-b')
plt.plot(t, zc, '-g')
plt.suptitle('Tensão Trifásica - abc', fontsize=30)
plt.ylabel('tensão(v)', fontsize=20)
plt.xlabel('tempo(t)', fontsize=20)
plt.legend(['fase - a', 'fase - b', 'fase - c'])
plt.arrow(-4, 0, 30, 0)
plt.arrow(0, -1.5, 0, 3)
plt.show()
```

```python
z_alpha = 1 * za - (1/2) * zb - (1/2) * zc 
z_beta  = 0 * za + (np.power(3, 1/2)/2) * zb - (np.power(3, 1/2)/2) * zc 

fig = plt.figure(figsize=(20, 6))
plt.suptitle('transformada abc - αβ', fontsize=30)
plt.plot(t, z_alpha, '-b')
plt.plot(t, z_beta, '-r')
plt.ylabel('tensão(v)', fontsize=20)
plt.xlabel('tempo(t)', fontsize=20)
plt.arrow(-4, 0, 30, 0)
plt.arrow(0, -2, 0, 4)
plt.legend(["alpha", "beta"])
plt.show() 
```

```python
teta = 90 * PI / 180
zd = z_alpha * np.cos(teta) + z_beta * np.sin(teta)
zq = - z_alpha * np.sin(teta) + z_beta * np.cos(teta)

fig = plt.figure(figsize=(20, 6))
plt.suptitle('transformada αβ- dq', fontsize=30)
plt.plot(t, zd, '-b')
plt.plot(t, zq, '-r')
plt.ylabel('tensão(v)', fontsize=20)
plt.xlabel('tempo(t)', fontsize=20)
plt.arrow(-4, 0, 30, 0)
plt.arrow(0, -2, 0, 4)
plt.legend(["eixo - d", "eixo - q"])
plt.show()
```
