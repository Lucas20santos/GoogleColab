```python
import numpy as np
import matplotlib.pyplot as plt
```

# Criando uma constante Chamada PI

```python
PI = np.pi
print(PI)
print(2 * PI)
print(2 * PI * 60)
```

```python
t = np.arange(- PI , 4 * PI + 0.0001, 0.0001)
w = 2 * PI * 60
za = np.cos(1 * t)
zb = np.cos(1 * t + (2 * PI / 3))
zc = np.cos(1 * t - (2 * PI / 3))
```

```python
fig = plt.figure(figsize=(20, 6))

plt.plot(t, za, '-r')
plt.plot(t, zb, '-b')
plt.plot(t, zc, '-g')
plt.suptitle('Tensão Trifásica - abc', fontsize=30)
plt.ylabel('tensão(v)', fontsize=20)
plt.xlabel('tempo(t)', fontsize=20)
plt.arrow(-4, 0, 30, 0)
plt.arrow(0, -1.5, 0, 3)
plt.show()
```

```python
z_alpha = 1 * za - (1/2) * zb - (1/2) * zc 
z_beta  = 0 * za + (np.power(3, 1/2)/2) * zb - (np.power(3, 1/2)/2) * zc 

fig = plt.figure(figsize=(20, 6))
plt.legend(["alpha", "beta"])
plt.suptitle('transformada abc - αβ', fontsize=30)
plt.plot(t, z_alpha, '-b')
plt.plot(t, z_beta, '-r')
plt.ylabel('tensão(v)', fontsize=20)
plt.xlabel('tempo(t)', fontsize=20)
plt.arrow(-4, 0, 30, 0)
plt.arrow(0, -2, 0, 4)
plt.show() 
```

```python
teta = 90 * PI / 180
zd = z_alpha * np.cos(teta) + z_beta * np.sin(teta)
zq = - z_alpha * np.sin(teta) + z_beta * np.cos(teta)

fig = plt.figure(figsize=(20, 6))
plt.legend(["d", "q"])
plt.suptitle('transformada αβ- dq', fontsize=30)
plt.plot(t, zd, '-b')
plt.plot(t, zq, '-r')
plt.ylabel('tensão(v)', fontsize=20)
plt.xlabel('tempo(t)', fontsize=20)
plt.arrow(-4, 0, 30, 0)
plt.arrow(0, -2, 0, 4)
plt.show()
```
